# Exercício 1

```python
import random
import numpy as np
import matplotlib.pyplot as plt

seed = 1000
```

```python
# AND
training_inputs = [[-1, -1],
                   [-1, 1],
                   [1, -1],
                   [1, 1]]

training_outputs = [-1, -1, -1, 1]
alfa = 0.05
```

```python
def train(t_inputs, t_outputs):
    alfa = 0.05
    weights = [random.random() for i in range(len(t_inputs[0]) + 1)]
    delta_w = [0 for i in range(len(weights))]
    iterations = 0
    error = True
    while error and iterations < 20000:
        error,weights = epoch(t_inputs, t_outputs, weights, delta_w, alfa)
        iterations = iterations + 1
    return iterations,weights


def epoch(t_inputs, t_outputs, w, d_w, a):
    outputs = []
    for i in range(len(t_inputs)):
        output = w[0] + t_inputs[i][0] * w[1] + t_inputs[i][1] * w[2]
        if output > 0:
            output = 1
        else:
            output = -1
        error = t_outputs[i] - output
        for j in range(len(d_w)):
            if j == 0:
                d_w[j] = d_w[j] + a * error
            else:
                d_w[j] = d_w[j] + a * t_inputs[i][j - 1] * error
        outputs.append(output)
    for i in range(len(w)):
        w[i] = w[i] + d_w[i]

    return not t_outputs == outputs,w
```

```python
runs = []
random.seed(seed)
for i in range(30):
    result=train(training_inputs, training_outputs)
    runs.append(result[0])
print('Average: ', sum(runs) / 30)
print('Standard deviantion: ', np.std(runs))
print('Max epochs: ', max(runs), '- Min epochs ', min(runs))
plt.boxplot([runs,np.std(runs),sum(runs) / 30])
labels = ['Runs','Standard-deviation','AVG Epochs']
plt.xticks([1,2,3],labels)
plt.show()
```
